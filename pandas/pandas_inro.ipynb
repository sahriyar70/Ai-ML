```python
import pandas as pd


left = pd.DataFrame({
    "key1": ["K0", "K1", "K2", "K3"],
    "A": [9, 5, 9, 6],
    "B": [1, 6, 9, 2]
})


right = pd.DataFrame({
    "key1": ["K0", "K2", "K4", "K5"],
    "C": [4, 9, 4, 4],
    "D": [3, 9, 4, 8]
})

print("Left:")
print(left)

print("\nRight:")
print(right)
```

```python
pd.merge(left,right , on='key1',)
```

```python
pd.merge(left,right, on='key1', how='outer')
```
