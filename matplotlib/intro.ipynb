```python
import numpy as np 
```

```python
x = np.arange(5,20)
x
```

```python
y = x*5 +1 
y 
```

```python
import matplotlib.pyplot  as plt 
```

```python
plt.plot(x,y, 'g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib graph ')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```
