factorial 

```python
def factorial (n):
    result = 1
    for i in range(1,n+1):
        result *= i 

    return result 

```

```python
factorial(5)
```

string revers 

```python
def str_revers (text):
    
    return text[::-1]
```

```python
str_revers("sahriiyar")
```

count vowel

```python
def count_vowel(text):
    vowel = "aeoiuAEIOU"
    count = 0  
    for char in text :
        if char in vowel :
            count += 1
    return count         
```

```python
count_vowel("ieidfdHHU")
```

prime number 

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

 
         
          
```

```python
print(is_prime(4))
```
