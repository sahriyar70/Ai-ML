leap year check 

```python
year = 2020 

if year % 4 == 0 and year % 400 == 0:
    print("leap year")
elif year % 4 == 0 and year % 100 != 0:
    print("leap year")   
else:
    print("not a leap year")    
```

simple calculator 

```python
a = 10 
b = 20 
operation = '*' 

if operation == '+':
    print(a + b)    
elif operation == '-':    print(a - b)
elif operation == '*':    print(a * b)   
elif operation == '/':    print(a / b)
else:    print("invalid operation")
```
